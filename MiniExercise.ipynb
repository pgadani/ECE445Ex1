```python
import numpy
from matplotlib import pyplot as plt

years = list(range(1930, 2011, 10))
gdp = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
```

```python
plt.figure()
plt.plot(years, gdp, 'r.--')
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.show()
```

The GDP curve 


The curve is roughly quadratic (the slope increases mostly increases steadily but it tapers off for the last data point). It seems like it could be centered at 1930 (the minimum known point).

It's easier to consider the quadratic in terms of $x-1930$, giving us $y=a(x-1930)^2+c$

At the minimum point, the value is close to $1$, so the constant term $c = 1$

In 2010, the value should be close to 16: $16 = a(80)^2+1$. The calculated $a=\frac{3}{1280}$ 

```python
xval = list(range(1930, 2011, 1))
yval = [3/1280*(x-1930)**2+1 for x in xval]

plt.figure()
plt.plot(years, gdp, 'r.--')
plt.title('Real US GDP (in trillions)')
plt.xlabel('Year')
plt.ylabel('Trillions of $')
plt.plot(xval, yval, 'b')
plt.legend(['GDP', 'estimated curve'])
plt.show()
```
